```python
from category_encoders import OrdinalEncoder
```

```python
data = pd.read_csv("loan_data.csv")
```

```python
categorical_features = [col for col in data.columns if data[col].dtypes == 'object']
encoder = OrdinalEncoder(cols=categorical_features).fit(data)
data2 = encoder.transform(data)
```

```python
data2.drop(['credit.policy', 'purpose', 'int.rate', 'days.with.cr.line'], axis=1, inplace=True)
```

```python
installmentFig = px.box(data2, y="installment", points="all")
installmentFig.show()
```

```python
annualLogFig = px.box(data2, y="log.annual.inc", points="all")
annualLogFig.show()
```

```python
dtiFig = px.box(data2, y="dti", points="all")
dtiFig.show()
```

```python
from re import X
ficoFig = px.histogram(data2, x="fico")
ficoFig.show()

fico = data2['fico']
fico[fico >= 800] = 5
fico[fico >= 740] = 4
fico[fico >= 670] = 3
fico[fico >= 580] = 2
fico[fico >= 300] = 1
data2['fico'] = fico

revolUtilFig2 = px.histogram(data2, x="fico")
revolUtilFig2.show()
```

```python
revolBalFig = px.box(data2, y="revol.bal", points="all")
revolBalFig.show()

revolBal = data2['revol.bal']

revolBalq1, revolBalq3 = np.percentile(revolBal, [25, 75])
revolBaliqr = revolBalq3 - revolBalq1
revolBalupper_fence = revolBalq3 + (1.5*revolBaliqr)


revolBal[revolBal >= revolBalupper_fence] = revolBalupper_fence
data2['revol.bal'] = revolBal

revolBalFig = px.box(data2, y="revol.bal", points="all")
revolBalFig.show()
```

```python
revolUtilFig = px.box(data2, y="revol.util", points="all")
revolUtilFig.show()

revilUtil = data2['revol.util']
revilUtil[revilUtil > 100] = 100
data2['revol.util'] = revilUtil

revolUtilFig2 = px.box(data2, y="revol.util", points="all")
revolUtilFig2.show()
```
